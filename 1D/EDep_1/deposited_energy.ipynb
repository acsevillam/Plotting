```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
x1, y1_1, y1_1_err, y1_2, y1_2_err = np.loadtxt('deposited_energy.txt',delimiter="\t",usecols =(0, 1, 2, 3, 4),unpack=True) 
```

```python
fig=plt.figure(figsize=(6,6))
#fig.suptitle("title", fontsize=16)

x_inter = np.linspace(10, 200, num=190, endpoint=True)

ax1=fig.add_subplot(111)
ax1.set_xlabel('Incident proton energy [keV]', fontsize=13,labelpad=15)
ax1.set_ylabel('Total deposited energy [keV]', fontsize=13,labelpad=15)
ax1.set_xlim([0, 210])
ax1.set_ylim([0, 210])
ax1.grid(linestyle='--', linewidth=0.25)

# 0 mg/ml (Reference)
pl=ax1.errorbar(x1, y1_1, yerr=4*y1_1_err, label='0 mg/ml', linewidth=0, marker='.', color="black", ms=5)
f1_1=interp1d(x1, y1_1, kind='linear')
pl=ax1.plot(x_inter, f1_1(x_inter), '--', linewidth=1, color="black")

# 20 mg/ml
pl=ax1.errorbar(x1, y1_2, yerr=4*y1_2_err, label='20 mg/ml', linewidth=0, marker='.', color="#440154FF", ms=5)
f1_2=interp1d(x1, y1_2, kind='linear')
pl=ax1.plot(x_inter, f1_2(x_inter), '--', linewidth=1, color="#440154FF")

ax1.legend(loc='best')
```
