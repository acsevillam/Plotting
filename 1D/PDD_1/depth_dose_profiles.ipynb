```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
x1, y1_1, y1_2, y1_3, y1_4, y1_5, y1_6, y1_7 = np.loadtxt('depth_dose_profiles.txt',delimiter="\t",usecols =(0, 1, 2, 3, 4, 5, 6, 7),unpack=True) 
```
